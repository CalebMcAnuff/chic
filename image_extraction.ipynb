```python
import pandas as pd
import requests
```

```python
# Load CSV file for downloading all images and saving them with their ID, followed by image extension
csv_path = "C:/Users/Mohsi/school/Winter2025/New folder/chic/chic_fashion_data.csv" 
df = pd.read_csv(csv_path)
```

```python
# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Referer": "https://www.aritzia.com/",
}
```

```python
# Try to Download a single image

# URL of the image
url = "https://assets.aritzia.com/image/upload/medium/s23_19_n08_103243_1274_on_a.jpg"

# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Referer": "https://www.aritzia.com/",
}

# Send GET request
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    with open("image.jpg", "wb") as file:
        file.write(response.content)
    print("Image downloaded successfully as 'image.jpg'")
else:
    print(f"Failed to download image. Status code: {response.status_code}")
```

```python
# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    image_url = row['Image URL']
    image_id = row['ID']


    
    # Send GET request
    response = requests.get(image_url, headers=headers)
    
    # Check if the request was successful
    if response.status_code == 200:
        with open(f"{image_id}.jpg", "wb") as file:
            file.write(response.content)
        print(f"Image {image_id}.jpg downloaded successfully.")
    else:
        print(f"Failed to download image {image_id}. Status code: {response.status_code}")

```
